# Modèle SIR, stratégie optimale de confinement

```python
import numpy as np
import matplotlib.pyplot as plt  
from scipy.integrate import odeint
from gekko import GEKKO
import warnings
m = GEKKO()
```

Dans ce T.P., nous cherchons à déterminer une stratégie optimale de mise en place de mesures de
restriction sur une population afin de lutter contre la propagation d’une épidémie. La question
que l’on souhaite résoudre s’énonce ainsi :

**Comment optimiser l’action des pouvoirs publics pour limiter le nombre de personnes infectées ?**

Nous allons considérer le modèle SIR (Susceptible-Infected-Removed) simplifié décrit dans [2] :

$$\left\{\begin{array}{ll} 
      S'(t) =-u(t) \frac{S(t)I(t)}{N(t)} \quad t \geq 0 \\
      I'(t) =u(t) \frac{S(t)I(t)}{N(t)}  - \beta I(t), \quad  \quad \quad (1) \\
      R'(t) =\beta I(t), \\
      (S(0), I(0), R(0)) = (S_0,  I_0, R_0).
      \end{array}
    \right.$$

Dans ce modèle $S(t)$ représente le nombre de personnes saines au temps t, $I(t)$ le nombre de
personnes infectées, et R(t) le nombre de personnes retirées (guéries ou décédées). La fonction
$u(·)$ représente un contrôle modélisant une politique de restriction, typiquement un confinement.
Il s’agit d’une fonction de $L^{\infty}(R^+)$ à valeurs dans $[0, \alpha]$. 
* Le cas $u(·) = 0$ modélise un confinement total 
* tandis que $u(·) = \alpha$ modélise l’absence de confinement. 
* Le nombre réel positif $\alpha$ représente le taux d’infection de la maladie. 
* Enfin, le nombre positif $\beta$ est le taux d’élimination.

La quantité $N = S(t) + I(t) + R(t)$ représente alors la population constante totale au cours du
temps. Il convient de bien différencier les personnes saines des personnes retirées : les personnes
saines n’ont pas encore été touchées par le virus, alors que les personnes retirées sont guéries,
et donc immunisées ou décédées. Autrement dit, les personnes retirées ne sont plus prises en
compte. Par conséquent, le modèle SIR ne s’occupe pas directement de prédire la mortalité de
l’épidémie.

Dans ce qui suit, nous allons nous baser sur la situation de l’épidémie en France en avril 2020.
Les auteurs de [1] estiment les paramètres $\alpha \simeq 0.32$ et $\beta = \frac {1}{10}$.

Rappelons que le taux de reproduction du virus est défini par $R_0 = \frac {\alpha} {\beta} \simeq 3.2$.

Il est bien connu que pour ce modèle, puisque $R_0 > 1$, $I$ atteint son maximum puis passe à
zéro en raison de l’immunité de groupe. Considérons le cas de la France, où le nombre total
d’individus est $N \simeq 6.7e7$. Nous allons considérer une fenêtre de temps $[0, T]$ avec $T = 90$ jours.

Nous utiliserons les valeurs numériques suivantes dans ce TP :

$$
\begin{array}{|l|c|c|c|c|c|}
\hline
\text{$\alpha$} & \beta & S_0 & I_0& R_0 & T \\
\hline
\text{0.32} & 0.1 & 6.7e7 - 2e4 & 2e4 & 0 &90 \\
\hline
\end{array}
$$

Nous choisissons de modéliser la recherche d’une stratégie de restriction optimale de la façon
suivante :

$$
\inf_{u \in \mathcal{U_{\alpha}}} L_{\tau}(u) \quad \quad avec \quad \mathcal{U_{\alpha}}=\{ u\in L^\infty(]0,T[])\mid 0\leq u(\cdot)\leq \alpha \} \quad \quad \quad (\mathcal P_{\tau})
$$
où $\tau \in [0.1]$ est fixé et 
$$
 L_{\tau}(u) = \tau \int_0^T g(u(t))dt + (1 - \tau) \int_0^T f(I(t)) dt
$$

La fonction g est supposée $C^1$, positive et décroissante sur $[0, \alpha]$ tandis que la fonction f est
supposée C^1, positive et croissante. On pourra par exemple choisir:
$$
g(u) = \frac 1 2 (\alpha^2 - u^2), \quad \quad f(I) = \frac{I^2}{2} \quad ou \quad f(I) = I.
$$

## Question 1:  Modélisation

**Justifier le choix de la fonctionnelle $L_τ$ .**

Comme le modèle SIR ne s’occupe pas directement de prédire la mortalité de
l’épidémie, l'enjeux peut être exprimé en deux points:

* Avoir une absence de confinement ($u(T) = \alpha$, d'où le choix d'une fonction décroissante positive telle g).
* Tout en ayant le but d'avoir un nombre minimal d'infectés ($I(T)=0$). 

Il est bien connu que pour ce modèle, puisque $R_0 > 1$, $I$ atteint son maximum puis passe à
zéro en raison de l’immunité de groupe. Alors, prendre une fonction croissante f est un choix judicieux.

## Question 2: N(.)

**Vérifier que la fonction N(·) est constante.**

le modéle est représenté par le système:
$$\left\{\begin{array}{ll} 
      S'(t) =-u(t) \frac{S(t)I(t)}{N(t)} \quad t \geq 0 \\
      I'(t) =u(t) \frac{S(t)I(t)}{N(t)}  - \beta I(t), \quad  \quad \quad (1) \\
      R'(t) =\beta I(t), \\
      (S(0), I(0), R(0)) = (S_0,  I_0, R_0).
      \end{array}
    \right.$$
Pour vérifier que $N(.)$, le nombre totale de la population, est constant on calcule $N'(.)$ à partir du systéme en dessus.

$$
N'(t) = S'(t) + I'(t) + R'(t) =
{-u(t) \frac{S(t)I(t)}{N(t)}} 
+ {u(t) \frac{S(t)I(t)}{N(t)}}  - {\beta I(t)}
+{\beta I(t)} = 0 \quad \quad \forall t \geq 0
$$

On en déduit que $N(.)$ est bien constante.

## Question 3: Cas de contrôles constants

**Dans chacun des deux cas suivants, représenter les solutions
S, I et R du système sur [0, T] :  $u(·) = α$ (pas de confinement), $u(·) = ε$ (avec $ε > 0$ proche de 0, confinement fort)**

On rappelle que pour résoudre l’EDO $x' = F(t, x)$, la méthode de Runge-Kutta d’ordre 4 consiste à utiliser
le schéma:
$x_{n+1} = x_n +\frac {h_n} 6(k_1 + 2k_2 + 2k_3 + k_4)$  avec:

$$\left\{\begin{array}{ll} 
      k_1 = F(t_n, x_n) \\
      k_2 = F(t_n + \frac {h_n} 2, x_n + h_n \frac {k_1} 2) \\
      k_3 = F(t_n + \frac {h_n} 2, x_n + h_n \frac {k_2} 2), \\
      k_4 = F(t_n + h_n, x_n + h_n k_3)).
      \end{array}
    \right.$$





**Et F: (t,x)->x' est implémenté ci-dessous:**

```python
def F(i,x,u,beta=0.1,N=6.7e7):
    assert len(x)==3,"Dimension de x=[S,I,R] incorrecte!"
    return np.array([-u[int(i)]*x[0]*x[1]/N, u[int(i)]*x[0]*x[1]/N - beta*x[1],beta*x[1]])
    
```

```python
def runge_kutta_4(F,t, x0,u,beta):

    dim_t = len(t) # temps
    dim_x = len(x0) # S,I et R (3)
    X = np.zeros((dim_t, dim_x))
    h_vec = t[1:] - t[:-1]#dans le cas ou h variable
    #N = x0[0]+x0[1]+x0[2]
    
    # initialisation
    X[0,:] = x0
    N= x0.sum()

    for i in range(0, dim_t-1):
        
        h = h_vec[i]#pas
        N = X[i,0]+X[i,1]+X[i,2]#vérifier que n est constante

        k1 = F(i, X[i,:],u=u,beta=0.1,N=N)
        k2 = F(i + h/2, 0.5*(X[i,:]+X[i-1,:]) + h*k1/2,u=u,beta=0.1,N=N)
        k3 = F(i + h/2, 0.5*(X[i-1,:]+X[i-1,:])  + h*k2/2,u=u,beta=0.1,N=N)
        k4 = F(i + h, X[i,:] + h*k3,u=u,beta=0.1,N=N)
        X[i+1,:] = X[i,:] + h/6 * (k1 + 2*k2 + 2*k3 + k4)
        

    return X
```

```python
def test_rungekutta_4(F,t,x0,u,beta=0.1,plot=True):
    with warnings.catch_warnings():
        warnings.filterwarnings('error')
        #np.seterr('raise')
        try:
            Sol = runge_kutta_4(F,t, x0,u,beta)
            if plot:
                print("-----------------------------------------------------------------------------------")
                print("Au bout T = ",T," jours (u=",u[-1],") la variables d'état [S,I,R] = \n", Sol[-1])
                print("-----------------------------------------------------------------------------------")
                fig, ax = plt.subplots(figsize=(9,6))
                ax.set(xlabel='Temps(jours)', ylabel='S,I,R,U', \
                       title= "Trajectoire pour une valeur de contôle constante ="+str(u[-1]))
                ax.plot(t,Sol[:,0],label=r"S")
                ax.plot(t,Sol[:,1],label=r"I")
                ax.plot(t,Sol[:,2],label=r"R")
                ax.plot(t,Sol[:,0]+Sol[:,1]+Sol[:,2],label=r"N")
                ax.plot(t,u*np.ones(len(t)),label=r"u")
                ax.grid(True)
                ax.legend(loc='best',fontsize=15);
            return Sol
        except Warning as e:
            print('error found:', e)
    
    
```

**Cas où le contôle $u=\alpha = 0.32$**

```python
T=90
nb = 1000
t=np.linspace(0,T,nb)
x0 = np.array([6.7e7-2e4,2e4,0])#S0,I0,R0
u = 0.32*np.ones(nb)

Sol = test_rungekutta_4(F,t,x0,u,plot=True)
```

**Commentaire:**

Comme attendue, le nombre d'infectés I(T) diminue aprés avoir la valeurs maximal, ceci est expliqué du fait que les infectés sont immunisées en partie ($\beta$). En plus la population suscéptible S(T) diminue aussi, ce qui peux être un état d'absence de confinement.

**Cas où le contôle $u= \epsilon$ avec $\epsilon \sim 0$:**

```python
T=90
nb = 1000
t=np.linspace(0,T,nb)
x0 = np.array([6.7e7-2e4,2e4,0])#S0,I0,R0

U = [0.2,0.15,0.1]
for u in U:
    u = u*np.ones(nb)

    Sol = test_rungekutta_4(F,t,x0,u,plot=True)
```

**Commentaire:**

Le nombre d'infectés I(T) est en hausse et visiblement loin de son maximum, alors que la population susceptible est trés grande ce qui engendre un confinement total.

## Question 4:  Calcul du gradient. 

Démontrer que la différentielle $DL_τ(u)·h$ du critère d’écrit:
$$
DL_{\tau}(u).h = lim_{\epsilon \rightarrow 0} \frac {L_{\tau}(u+\epsilon h).h - L_{\tau}(u).h}{\epsilon}
= 
\int_0^T h\big( \tau g'(u) + (1-\tau) \frac {S I}{N}(P_2 - P_1)\big)
$$

où P désigne l'état adjoint associé à ce problème, solution du système:

$$ 
\left\{\begin{array}{ll} 
\frac {d} {dt}\begin{bmatrix}
    P_1 \\
    P_2  \\
    P_3
    \end{bmatrix} 
    = A^{T}
    \begin{bmatrix}
    P_1 \\
    P_2  \\
    P_3
    \end{bmatrix} 
    +
    \begin{bmatrix}
    0 \\
    f'(I)  \\
    0
    \end{bmatrix}, \quad \quad \quad A = \begin{bmatrix}
                                                - \frac {uI} {N}  & - \frac {uS} {N} & 0\\
                                                 \frac {uI} {N} &  \frac {uS} {N} - \beta & 0 \\
                                                0 & \beta & 0
                                                \end{bmatrix} \\
      P_i(T) = 0 ,
      \end{array}
    \right.
$$


Comme indiqué dans `l'aide complémentaire` la différentiabilité du critère $L_{\tau}$ sera admise (elle repose sur un argument type "fonctions implicite"). On calcule alors la différentielle, et on procède en 3 étapes:

**Étape 1: première expression de la différentielle:**

Soit $u ∈ \cal U_α$ et $h ∈ L^∞(]0, T[)$, une perturbation admissible de u
dans $\cal U_α$, autrement dit telle que $u + εh ∈ \cal U_α$ si $ε > 0$ est assez petit.

On a alors:

$$
DL_{\tau}(u).h = lim_{\epsilon \rightarrow 0} \frac {L_{\tau}(u+\epsilon h) - L_{\tau}(u)}{\epsilon}
=
\frac {d} {d\epsilon} L_{\tau}(u+\epsilon h)\mid_{\epsilon =0} 
$$

Soit $(S',I',R')$ la différentielle de $(S_u,I_u,R_u)$ en u dans la direction h (où $(S_u,I_u,R_u)$) la solution du système $(1)$ associée au choix de contôle u), autrement dit:

$$
S'(.) = \frac {dS_{u+\epsilon h}}{d \epsilon}\mid_{\epsilon=0} \\
I'(.) = \frac {dI_{u+\epsilon h}}{d \epsilon}\mid_{\epsilon=0} \\
R'(.) = \frac {dR_{u+\epsilon h}}{d \epsilon}\mid_{\epsilon=0} 
$$

On a:

$$
DL_{\tau}(u).h 
=
\frac {d L_{\tau}(u+\epsilon h)\mid_{\epsilon =0} } {d\epsilon} 
$$
On rappel que:
$$
 L_{\tau}(u) = \tau \int_0^T g(u(t))dt + (1 - \tau) \int_0^T f(I(t)) dt
$$

Donc:

$$
 L_{\tau}(u+\epsilon h) = \tau \int_0^T g(u(t)+\epsilon h)dt + (1 - \tau) \int_0^T f(I_{u+\epsilon h}) dt
$$

Puisque $g$ et $f$ sont de classe $C^1$, aussi $\epsilon \rightarrow u+\epsilon h$ et $\epsilon \rightarrow I_{u+\epsilon h}$ on applique la régle de chaine ($D_a(f o g) = (D_{g(a)f}) o D_a g)$)

$$
DL_{\tau}(u).h 
=
\tau \int_0^T \frac {d g(u+\epsilon h)\mid_{\epsilon =0} } {d\epsilon} dt + (1 - \tau) \int_0^T \frac {d f(I_{u+\epsilon h}(t))\mid_{\epsilon =0} } {d\epsilon} dt
$$

Alors (on posant f' , g' les dérivées réspective de f et g)

$$
DL_{\tau}(u).h 
=
\tau \int_0^T g'(u+\epsilon h)\mid_{\epsilon =0}\frac {d(u+\epsilon h)\mid_{\epsilon =0} } {d\epsilon} dt + (1 - \tau) \int_0^T f'(I_{u+\epsilon h})\mid_{\epsilon =0} \frac {d(I_{u+\epsilon h})\mid_{\epsilon =0} } {d\epsilon} dt
$$

On évalue en $\epsilon = 0$:

En effet : $\frac {d(u+\epsilon h){}\mid_{\epsilon =0} } {d\epsilon} = h(t)$

et $\frac {d(I_{u+\epsilon h})\mid_{\epsilon =0} } {d\epsilon} = I'(t)$

$$
DL_{\tau}(u).h 
=
\tau \int_0^T g'(u(t))h(t) dt + (1 - \tau) \int_0^T f'(I_{u}(t))I'(t)dt \quad \quad (*)
$$

Maintenant il faut déterminer $I'$ pour la remplacer dans $(*)$ et aussi $S'$ et $R'$ au passage.

**Étape 2: détermination de (S^*,I^*,R^*):**

On a $\forall t \geq 0$:

$$
S'_{u(t)+\epsilon h(t)}(t) = -(u(t)+\epsilon h(t))\frac {S_{u(t)+\epsilon h(t)} I_{u(t)+\epsilon h(t)}}{N}
$$

Dérivons cette relation par rapport à $\epsilon$ 

**Pour ce qui suit on symbolise la dérivation par rapport $\epsilon$ par * pour ne pas confondre avec ' qui est par rapport à t**

en effet par abus de notation:

$$
S^*(.) = \frac {dS_{u+\epsilon h}}{d \epsilon}\mid_{\epsilon=0} \\
I^*(.) = \frac {dI_{u+\epsilon h}}{d \epsilon}\mid_{\epsilon=0} \\
R^*(.) = \frac {dR_{u+\epsilon h}}{d \epsilon}\mid_{\epsilon=0} 
$$

on a donc: 

$$
S'^*_{u(t)+\epsilon h(t)}(t) = - \big((u(t)+\epsilon h(t))\frac {(S_{u(t)+\epsilon h(t)} I_{u(t)+\epsilon h(t)})}{N}\big)^*
$$

$ \iff$ ((f.g)' = f'g +fg' et on rappel que N est constante) 

$$
S'^*_{u(t)+\epsilon h(t)}(t) = -(u(t)+\epsilon h(t))^*\frac {(S_{u(t)+\epsilon h(t)} I_{u(t)+\epsilon h(t)})}{N}
-(u(t)+\epsilon h(t))\big(\frac {(S_{u(t)+\epsilon h(t)} I_{u(t)+\epsilon h(t)}}{N}\big)^*
$$

$ \iff$ $((u(t)+\epsilon h(t))^* = h(t))$

$$
S'^*_{u(t)+\epsilon h(t)}(t) = - h(t)\frac {(S_{u(t)+\epsilon h(t)} I_{u(t)+\epsilon h(t)})}{N}
-(u(t)+\epsilon h(t))\frac {(S^*_{u(t)+\epsilon h(t)} I_{u(t)+\epsilon h(t)})}{N}
-(u(t)+\epsilon h(t))\frac {(S_{u(t)+\epsilon h(t)} I^*_{u(t)+\epsilon h(t)})}{N}
$$

On utilisons les notations convenue:

$$
S^*(.) = S^*_{u+\epsilon h} = \frac {dS_{u+\epsilon h}}{d \epsilon}\mid_{\epsilon=0} \\
I^*(.) = I^*_{u+\epsilon h} = \frac {dI_{u+\epsilon h}}{d \epsilon}\mid_{\epsilon=0} 
$$
On permute ' et * et on évalue l'égalité obtenue en $\epsilon = 0$ 

$$
S^*(t)' = - h(t)\frac {(S_{u(t)} I_{u(t)})}{N}
-u(t)\frac {(S^*(t) I_{u(t)})}{N}
-u(t)\frac {(S_{u(t)} I^*(t))}{N}
$$

De la même façon:

$$
I^*(t)' =  h(t)\frac {(S_{u(t)} I_{u(t)})}{N}
+u(t)\frac {(S^*(t) I_{u(t)})}{N}
+u(t)\frac {(S_{u(t)} I^*(t))}{N}
- \beta I^*(t)
$$

et

$$
R^*(t)' = \beta I^*(t)
$$

Avec $(S^*(0),I^*(0),R^*(0)) = (0,0,0)$

en effet:

$$
S^*(0) = \frac {dS_{u+\epsilon h}(0)}{d \epsilon}\mid_{\epsilon=0}  = lim_{\epsilon \rightarrow 0} \frac {S_{u+\epsilon h}(0) - S_{u}(0)}{\epsilon} = 0 \quad (Condition \, \, initiales)
$$

et pareil pour $I^*(0)$ et $ R^*(0)$

**Étape 3: calcul du gradient de $L_{\tau}$

En outre, on prends les différentieles de la variable de l'état selon la direction h $(S^*,I^*,R^*)$ et l'état adjoint du problème associé $(P_1,P_2,P_3)$ et on effectue la multiplication au sens du produit scalaire de ces dérniers avec les équations obtenues ci-dessus $$(S^*(t)',I^*(t)',R^*(t)')$$

$$
\int_0^T S^*(t)'P_1 dt + \int_0^T I^*(t)'P_2 dt + \int_0^T R^*(t)'P_3 dt

\int_0^T \big[- P_1 h(t)\frac {(S_{u(t)} I_{u(t)})}{N}

+P_2 h(t)\frac {(S_{u(t)} I_{u(t)})}{N}

+P_2 u(t)\frac {(S_{u(t)} I^*(t))}{N}

$$

                                                \end{bmatrix} \\
      P_i(T) = 0 ,
      \end{array}
    \right.
$$

On obtient donc:

$$
\int_0^T \begin{bmatrix}
    S^* \\
    I^*  \\
    R^*
    \end{bmatrix}
    \begin{bmatrix}
    0 \\
    f'(I)  \\
    0
    \end{bmatrix}dt= 
\int_0^T \frac {h(t)S_{u(t)} I_{u(t)}}{N} \big[P_2- P_1\big]dt 
$$

Cette idendité est appelé identité de dualité. Ainsi:

$$
\int_0^T I^*  f'(I) dt= 
\int_0^T \frac {h(t)S_{u(t)} I_{u(t)}}{N} \big[P_2- P_1\big]dt 
$$

Conséquent, en remplaçant dans $(*)$ on obtient:
$$
DL_{\tau}(u).h 
=
\tau \int_0^T g'(u(t))h(t) dt + (1 - \tau) \int_0^T \frac {h(t)S_{u(t)} I_{u(t)}}{N} \big[P_2- P_1\big]dt 
=(h,\nabla L_{\tau})_{L^2(]0,T[)}
$$
Posons alors: $$\psi(t) = \nabla L_{\tau} = \tau g'(u) + (1-\tau) \frac {S I}{N}(P_2 - P_1)$$ comme étant le gradient de la fonctionnelle $L_{\tau}$ p.p. $t \in [0,T]$

## Question 5: Mise en œuvre d’une méthode de gradient.

Dans la mise en oeuvre de la méthode de gradient présentée ci-dessous, on a suivis les étapes suivantes:

* utiliser la formule de projection
$$u^{k+1} = min\{α, max\{u^k − ρ^k∇L_τ (u^k), 0\}\}.$$
* le pas $\rho^k$ peut être choisi variable, suivant l’algorithme ci-après.

$\quad cpt = 0$ 

$\quad \rho =\rho_0 $ (un nombre positif fixé)

$\quad \Delta = {L_{\tau}(u^k − ρ^k∇L_τ (u^k))-L_{\tau}(u^k)}$

$\quad$ **while** $(cpt<1e2)$ & $(\Delta \geq 0)$

$\quad \rho \leftarrow \frac \rho {1.3} \quad$ 
&
$\quad \Delta \leftarrow {L_{\tau}(u^k − ρ^k∇L_τ (u^k))-L_{\tau}} $

$\quad cpt = cpt+1$

* on imposera un nombre maximal d’iterations. Si ce nombre est atteint, on considérera
que l’algorithme n’a pas convergé.

* On utilisera un critère d’arrêt de la forme $\mid{\frac {L_{\tau}(u^{k+}-L_{\tau}(u^k))}{L_{\tau}(u^k)}}\mid \leq \epsilon_{tol}$ où $ε_{tol} > 0$ est fixé.

**Remarque: dans l'implémentation j'ai pris f(I) = I**

**Calcul de l'adjoint**

Il est basé sur un bout de code que vous avez partagé avec mes camarades.

```python
##code que vous avez partagé avec mes camarades
## second terme du pb adjoint
def f_dual(i,u_i,ystate,p_i,cas=1):
    a11=-u_i*ystate[i,1]/N
    a12=-u_i*ystate[i,0]/N
    a21=u_i*ystate[i,1]/N
    a22=u_i*ystate[i,0]/N-beta
    A=np.matrix([[a11, a12, 0],[a21 ,a22, 0],[ 0 ,beta, 0]])
    if cas==1:
        return np.transpose(np.transpose(A)*np.transpose(np.matrix(p_i))+np.matrix([[0],[1],[0]]))
    else:
        return np.transpose(np.transpose(A)*np.transpose(np.matrix(p_i))+np.matrix([[0],[ystate[i,1]],[0]]))
        
def RK4_adj(u,ystate,NT,cas=1):
    h = T/NT
    pT = np.zeros((NT,3))
    for i in list(reversed(range(1,NT))):
        k1 = h*f_dual(i,u[i],ystate,pT[i,:],cas)
        k2 = h*f_dual(i,0.5*(u[i]+u[i-1]),ystate,pT[i,:]+0.5*k1,cas)
        k3 = h*f_dual(i,0.5*(u[i]+u[i-1]),ystate,pT[i,:]+0.5*k2,cas)
        k4 = h*f_dual(i,u[i-1],ystate,pT[i,:]+k3,cas)
        pT[i-1,:] = pT[i,:] + (k1 + 2*k2 + 2*k3 + k4)/6

    return pT
```

**Calcule de SIR: variable de l'état**

```python
#second terme du pb (1)
def F(i,x,u,beta=0.1,N=6.7e7):
    assert len(x)==3,"Dimension de x=[S,I,R] incorrecte!"
    return np.array([-u[int(i)]*x[0]*x[1]/N, u[int(i)]*x[0]*x[1]/N - beta*x[1],beta*x[1]])
```

```python
def runge_kutta_4(F,t, x0,u,beta):

    dim_t = len(t) # temps
    dim_x = len(x0) # S,I et R (3)
    X = np.zeros((dim_t, dim_x))
    h_vec = t[1:] - t[:-1]#dans le cas ou h variable
    #N = x0[0]+x0[1]+x0[2]
    
    # initialisation
    X[0,:] = x0
    N= x0.sum()

    for i in range(0, dim_t-1):
        
        h = h_vec[i]#pas
        #N = X[i,0]+X[i,1]+X[i,2]#vérifier que n est constante
        #print(N)
        k1 = F(i, X[i,:],u=u,beta=0.1,N=N)
        k2 = F(i + h/2, 0.5*(X[i,:]+X[i-1,:]) + h*k1/2,u=u,beta=0.1,N=N)
        k3 = F(i + h/2, 0.5*(X[i,:]+X[i-1,:])  + h*k2/2,u=u,beta=0.1,N=N)
        k4 = F(i + h, X[i,:] + h*k3,u=u,beta=0.1,N=N)
        X[i+1,:] = X[i,:] + h/6 * (k1 + 2*k2 + 2*k3 + k4)
        

    return X
```

**Normalisation:**

On prend la valeur la valeur qui correspond à $u = \frac {\alpha } {2}$. le calcule des intégrales se fait par la méthode des réctangles.

```python
def L_avg(SIR,u,cas = 1):
    L_g = 0
    L_f = 0
    I = SIR[:,1]
    h = T / nb
    for i in range(nb):
        L_g += tau*(alpha**2 - (alpha/2)**2)*h
        if cas==1:
            L_f += (1-tau)*I[i] * h
        else:
            L_f += (1-tau)*I[i]**2/2* h
        
    return L_g,L_f
```

**Calcule de $L_{\tau}$ (méthode des réctangles): Le critère normalisé**

```python
def L(SIR,L_g,L_f,cas=1):
    somme = 0
    h = T / nb
    I = SIR[:,1]
    for i in range(0, nb):
        if cas==1:
            somme += tau*(alpha**2 - u[i]**2)*h/L_g +(1-tau)*I[i]* h/L_f #+normalisation
        else:
            somme += tau*(alpha**2 - u[i]**2)*h/L_g +(1-tau)*I[i]**2/2 * h/L_f #+normalisation
        
    return somme
```

**Calcule du gradient de $L_{\tau}$:**

```python
def grad_L(u, SIR, P,L_g,L_f):
    S = SIR[:,0]
    I = SIR[:,1]
    P_1= P[:,0]
    P_2= P[:,1]
    return -tau*2*u/ L_g+ (1-tau)* (P_2 - P_1)* S*I/N/L_f
```

**Initialisation:**

```python
#liés au problème
T=90
nb = 100
t=np.linspace(0,T,nb)
x0 = np.array([6.7e7-2e4,2e4,0])#S0,I0,R0
N = 6.7e7
beta = 0.1
alpha = 0.32

u = np.ones(nb) * alpha 

tau = 0.5
```

**Méthode du gradient:**

```python
def gradient_SIR(cas =1,u=u):
    listeL = []
    """initialisation de SIR puis P et L (normalisé)"""
    SIR = runge_kutta_4(F,t, x0,u,beta)
    P = RK4_adj(u, SIR,nb,cas)
    L_g,L_f = L_avg(SIR,u,cas)#constantes de normalisation
    #print(L_g,L_f)

    listeL.append(L(SIR,L_g,L_f))
    """Résoudre inf_u L"""
    epstol = 1e-12
    err = 1
    rho = 1
    iteration=0
    iterMax = 100
    delta = 1
    while err > epstol and iteration < iterMax:
        """calculer rho """
        cpt = 0
        #print(L_g,L_f)
        gradient = grad_L(u, SIR, P,1,1)
        SIR2 = runge_kutta_4(F,t, x0,np.clip(u - rho * gradient, 0, alpha),beta)
        delta = L(SIR2,L_g,L_f,cas) - listeL[-1]
        while cpt < 100 and delta >= 0:
            rho /= 1.3
            SIR2 = runge_kutta_4(F,t, x0,np.clip(u - rho * gradient, 0, alpha),beta)
            delta = L(SIR2,L_g,L_f,cas) - listeL[-1]
            cpt += 1

        #print(cpt)

        """ mise à jour de u """
        np.clip(u - rho * gradient, 0, alpha, out=u)
        SIR = runge_kutta_4(F,t, x0,np.clip(u - rho * gradient, 0, alpha),beta)

        P = RK4_adj(u, SIR,nb,cas)
        listeL.append(L(SIR,L_g,L_f,cas))
        err = abs((listeL[-1] - listeL[-2]) / listeL[-2])
        iteration += 1
        #print(u)
        #print(L(SIR,L_g,L_f))
    print("on a fait ",iteration, "pour converger vers u optimale")
    return SIR,u,listeL,iteration
```

### Question (b): GEKKO

On rappel que nous choisissons de modéliser la recherche d’une stratégie de restriction optimale de la façon
suivante :

$$
\inf_{u \in \mathcal{U_{\alpha}}} L_{\tau}(u) \quad \quad avec \quad \mathcal{U_{\alpha}}=\{ u\in L^\infty(]0,T[])\mid 0\leq u(\cdot)\leq \alpha \} \quad \quad \quad (\mathcal P_{\tau})
$$
où $\tau \in [0.1]$ est fixé et 
$$
 L_{\tau}(u) = \tau \int_0^T g(u(t))dt + (1 - \tau) \int_0^T f(I(t)) dt
$$

La fonction g est supposée $C^1$, positive et décroissante sur $[0, \alpha]$ tandis que la fonction f est
supposée C^1, positive et croissante. On pourra par exemple choisir:
$$
g(u) = \frac 1 2 (\alpha^2 - u^2), \quad \quad f(I) = \frac{I^2}{2} \quad ou \quad f(I) = I.
$$

Nous allons considérer le modèle SIR (Susceptible-Infected-Removed) simplifié décrit dans [2] :

$$\left\{\begin{array}{ll} 
      S'(t) =-u(t) \frac{S(t)I(t)}{N(t)} \quad t \geq 0 \\
      I'(t) =u(t) \frac{S(t)I(t)}{N(t)}  - \beta I(t), \quad  \quad \quad (1) \\
      R'(t) =\beta I(t), \\
      (S(0), I(0), R(0)) = (S_0,  I_0, R_0).
      \end{array}
    \right.$$
    
**Préparation gekko**

* Les intégrales sont remplacé par des variables d'état G,F telle que :
$$
G ' = \tau  g(u(t)) \\
F' = (1 - \tau)f(I(t))
$$

* l'initialisation $u^0 = a$ est testé pour $a = [0.05,0.2,0.3199]$.

```python
for a in [0.05,0.2,0.3199]:  
    m = GEKKO(remote=False) # initialize GEKKO
    nt = 100
    m.time = np.linspace(0.1,90,100)
    N = 6.7e7
    beta = 0.1
    alpha =0.32
    tau = 0.5
    tm = m.time
    
    # Variables
    S = m.Var(value=6.7e7-2e4)
    I = m.Var(value=2e4)
    R = m.Var(value=0.0)
    #variables pour calcul de l'inégrale
    G = m.Var(value=0)
    F = m.Var(value=0)

    # control changes every time period
    u = m.MV(value=a,lb=0,ub=alpha)
    u.STATUS = 1
    m.Equation(S.dt()==-u*S*I/N)
    m.Equation(I.dt()==u*S*I/N - beta*I)
    m.Equation(R.dt()==beta*I)
    m.Equation(G.dt()==(alpha**2 - u**2)*tau)
    m.Equation(F.dt()==I**2/2*(1-tau))
    
    m.Obj((G+F)**2)

    m.options.IMODE = 6
    m.solve(disp=False)
    plt.figure(1)
    plt.plot(tm,S.value,'k-',label=r'$S$ pour $u^0=$'+str(a))
    plt.plot(tm,I.value,'b-',label=r'$I$ pour $u^0=$'+str(a))
    plt.plot(tm,R.value,'g--',label=r'$R$ pour $u^0=$'+str(a))
    plt.legend(loc='best')
    plt.xlabel('Time')
    plt.show()
    plt.figure(2)
    plt.plot(tm,u.value,'r--',label=r'$u$ pour $u^0=$'+str(a))
    plt.legend(loc='best')
    plt.xlabel('Time')
    plt.show()
    plt.figure(3)
    plt.plot(tm,G.value,'r--',label=r'$terme confinement$')
    plt.plot(tm,F.value,'v--',label=r'$terme infectés$')
    plt.legend(loc='best')
    plt.xlabel('Time')
    plt.show()
```

**Conclusion:**

* La valeur des termes g et f aboutissent qualitativement sur la même conclusion, en effet le terme infectationest bien dominant.

* Le contôle est bien bangbang.

* Les trajectoires optimales ont aussi un profile de confinemnt tôtal comme obtenue par la méthode des gradients.
